```python
links=[]
name=[]
mark=[]
episode=[]
year=[]
comms=[]
rating=[]
genre=[]
watched=[]
typee=[]
author=[]
#вводим массивы
from bs4 import BeautifulSoup
import requests 
import time
from fake_useragent import UserAgent
for p in range(1,26):
    url='https://shikimori.one/animes/page/{}.html'.format(p)#ходим по страницам сайта (25 страниц по 20 аниме на одной = 500 строк в df)
    response = requests.get(url, headers={'User-Agent': UserAgent().chrome})#используем фейк агент,тк без него код 403
    tree =  BeautifulSoup(response.content, 'html.parser')#создаём дерево
    source = tree.find_all('article')
    for i in range(20):#ходим по ссылка на сайте, на 1 старнице 20 ссылок
        links.append(source[i].find('a').get('href'))#добавляем в массив ссылок ссылки на страницы аниме
        time.sleep(0.1)#добавляем тайм слип, тк без него переодически вылазит код403 (до поломки кода тайм слип был не нужен)
        name.append(source[i].find('span',{'class':'name-ru'}).text)# добавляем названия аниме на русском
for lol in range(500):
    url_new=links[lol]#при каждом новом значении lol меняем url и ходим уже по страницам аниме
    response_new=requests.get(url_new, headers={'User-Agent': UserAgent().chrome})
    tree_new =  BeautifulSoup(response_new.content, 'html.parser')
    if tree_new.find('div', {'class':'score-value score-9'}) is None:#если оценка аниме меньше 9, то вместо 'score-value score-9' надо искать по 'score-value score-8'
        if tree_new.find('div', {'class':'score-value score-8'}) is None:
            mark.append('None')#сделано на случай непредвиденнызх обстоятельств (как поломка кода)
        else:
            mark.append(tree_new.find('div', {'class':'score-value score-8'}).text)
    else:
        mark.append(tree_new.find('div', {'class':'score-value score-9'}).text)#добавляем в массив mark пользовательскую оценку аниме
    if tree_new.find('div',{'class':'value'}) is not None:
        episode.append(tree_new.find_all('div', {'class':'line'})[1].find('div',{'class':'value'}).text)#добавляем количество эпизодов
    else:
        episode.append(1)
    typee.append(tree_new.find_all('div', {'class':'line'})[0].find('div',{'class':'value'}).text)#добавляем тип аниме(сериал,фильм,ова и тд)
    if tree_new.find_all('span')[15].get('title') is not None:
        year.append(tree_new.find_all('span')[15].get('title'))#добавляем год издания аниме
    else:
        year.append(tree_new.find_all('div',{'class':'value'})[2].text)#сайт написан немного криво, так что приходится импровизировать и находить дату в другом месте того же сайта
    genre.append(tree_new.find('span',{'class':'genre-ru'}).text)#добавляем жанр
    if tree_new.find('div',{'class':'additional-links'}) is not None:
        comms.append(tree_new.find('div',{'class':'additional-links'}).find_all('span')[-1].text)#добавляем количество комментариев
    else:
        comms.append('None')#на случай непредвиденных обстоятельств
#Код работал как часы до определённого момента(примерно 21:00 6 июня), пока не начал выдавать ошибки, при чём иногда разные. При этом никаких правок в код не вносилось. Попытки запустить через ТОР или ввести time.sleep() на каждом этапе не помогли. Код просто сломался и всё.
#Файл 'starting.csv' является последней доступной версией цсвэшника датафрейма, который получался из этого кода(точнее, более раннеё версии этого кода).
import pandas as pd
u={'name':name, 'link':links, 'mark':mark,'episode or len':episode,'type':typee,'genre':genre,'comments':comms,'year':year}
mu = pd.DataFrame(u)
mu.to_csv('starting.csv')#получаем csv файл, еслу нужно.

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
from statsmodels.robust.scale import mad
from matplotlib import rcParams
rcParams['figure.figsize'] = (20, 7)
rcParams['axes.titlesize'] = 20
df = pd.read_csv('C://Users//Admin//starting.csv', index_col=0)
df['mark']
 
df1=
```
