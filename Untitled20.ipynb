```python
links=[]
name=[]
mark=[]
episode=[]
year=[]
comms=[]
rating=[]
genre=[]
watched=[]
typee=[]
author=[]

from bs4 import BeautifulSoup
import requests 
import time
from fake_useragent import UserAgent
for p in range(1,26):
    url='https://shikimori.one/animes/page/{}.html'.format(p)
    response = requests.get(url, headers={'User-Agent': UserAgent().chrome})
    tree =  BeautifulSoup(response.content, 'html.parser')
    source = tree.find_all('article')
    for i in range(20):
        links.append(source[i].find('a').get('href'))
        name.append(source[i].find('span',{'class':'name-ru'}).text)
for lol in range(500):
    url_new=links[lol]
    response_new=requests.get(url_new, headers={'User-Agent': UserAgent().chrome})
    tree_new =  BeautifulSoup(response_new.content, 'html.parser')
    if tree_new.find('div', {'class':'score-value score-9'}) is None:
        mark.append(tree_new.find('div', {'class':'score-value score-8'}).text)
    else:
        mark.append(tree_new.find('div', {'class':'score-value score-9'}).text)
    if tree_new.find('div',{'class':'value'}) is not None:
        episode.append(tree_new.find_all('div', {'class':'line'})[1].find('div',{'class':'value'}).text)
    else:
        episode.append(1)
    typee.append(tree_new.find_all('div', {'class':'line'})[0].find('div',{'class':'value'}).text)
    year.append(tree_new.find_all('span')[15].get('title'))
    genre.append(tree_new.find('span',{'class':'genre-ru'}).text)

    
print(mark[:10])
print(mark[490:])
print(typee[:10])
print(typee[490:])
print(episode[:10])
print(episode[490:])
print(year[:10])
print(year[490:])
print(genre[:10])
print(genre[490:])
print(name[:10])
print(name[490:])
print(links[:10])
print(links[490:])
```
